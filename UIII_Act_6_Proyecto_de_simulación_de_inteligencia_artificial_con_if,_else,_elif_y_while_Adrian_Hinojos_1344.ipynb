```python
print("💳 BILLETERA DIGITAL INTELIGENTE")

saldo = 1000
opcion = ""
continuar = "si"

while continuar == "si":

    print("\n------ MENÚ ------")
    print("1. Consultar saldo")
    print("2. Depositar dinero")
    print("3. Retirar dinero")
    print("4. Recomendación inteligente")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # CONSULTAR SALDO
    if opcion == "1":
        print("Tu saldo actual es: $", saldo)

    # DEPOSITAR DINERO
    elif opcion == "2":

        deposito = int(input("¿Cuánto dinero deseas depositar?: "))
        saldo = saldo + deposito

        print("Depósito realizado correctamente.")
        print("Nuevo saldo: $", saldo)

    # RETIRAR DINERO
    elif opcion == "3":

        retiro = int(input("¿Cuánto dinero deseas retirar?: "))

        if retiro <= saldo:
            saldo = saldo - retiro
            print("Retiro exitoso.")
            print("Saldo restante: $", saldo)

        else:
            print("❌ Saldo insuficiente.")

    # IA SIMPLE
    elif opcion == "4":

        print("\n🤖 Analizando tus finanzas...")

        if saldo < 200:
            print("Recomendación: debes ahorrar más dinero.")

        elif saldo >= 200 and saldo <= 1000:
            print("Tus finanzas son estables.")

        elif saldo > 1000:
            print("Excelente ahorro. Puedes invertir parte de tu dinero.")

    # SALIR
    elif opcion == "5":

        print("Gracias por usar la billetera digital.")
        continuar = "no"

    # OPCIÓN INVÁLIDA
    else:
        print("⚠️ Opción no válida.")
```
